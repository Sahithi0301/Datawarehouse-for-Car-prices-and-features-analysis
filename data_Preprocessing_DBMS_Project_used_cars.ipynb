```python
# Drop Nulls Because The Full Row Is Null
df.dropna(inplace=True)
```

```python
# Split Car Full Data into Brand, Model, And Year of The Model

model = df['Car Model'].apply(lambda x: ''.join((x.split()[1],' ',x.split()[2])) if len(x.split()) >3 else x.split()[1])
year = df['Car Model'].apply(lambda x: x.split()[-1])
brand = df['Car Model'].apply(lambda x: x.split()[0])

df['Manufacturer'], df['Model'], df['ProdYear'] = brand,model,year
```

```python
# Convert Average price, Minimum price, Maximum price to Numaarical Values

for i in ['Average price','Minimum price','Maximum price']:
    df[i] = df[i].apply(lambda x : float((str(x).replace(',','')).split()[0]))
```

```python
columns_to_drop = ['Car Model', 'web-scraper-order']
df = df.drop(columns_to_drop, axis=1)
```

```python
# check for white spaces in string values in all columns
for col in df.columns:
    # convert non-string values to string
    mask = df[col].apply(lambda x: str(x)).str.contains(r'^\s|\s$')
   
    num_spaces = mask.sum()
    if num_spaces > 0:
      print(f'Column "{col}" contains {num_spaces} values with white spaces.')
    else:
      print("null")
```
