```python
import pandas as pd
```

```python
df=pd.read_csv('/content/cars_features.csv')
df.shape
df
```

```python
df.isnull().sum()
```

```python
columns_to_drop = ['Price', 'Levy','Doors','Wheel','Drive wheels']
df = df.drop(columns_to_drop, axis=1)
```

```python
df['Engine volume'] = df['Engine volume'].str.replace('Turbo', '')
```

```python
df
```

```python
for i in ['Mileage']:
    df[i] = df[i].apply(lambda x : float((str(x).replace(',','')).split()[0]))
```

```python
#df.to_csv("cars_df_cleaned.csv", index=False)
```

```python
sorted(df['Manufacturer'].unique())
```

```python
# Define a function to replace 0 with pd.NA
def replace_zero(value):
    return pd.NA if value == 0 else value

# Apply the function to each element in the DataFrame
df = df.applymap(replace_zero)

```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)

```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'Prod. year':'ProdYear',	'Leather interior':'Leather_interior', 'Fuel type':'Fuel_type', 'Engine volume':'Engine_volume','Gear box type':'Gear_box_type' })
```

```python
df
```

```python
df.dtypes
```

```python
df['Cylinders'] = df['Cylinders'].astype(int)
df['Engine_volume'] = df['Engine_volume'].astype(float)
df['Mileage'] = df['Mileage'].astype(int)
df['Airbags'] = df['Airbags'].astype(int)
df['Manufacturer'] = df['Manufacturer'].astype(str)
df['Category'] = df['Category'].astype(str)
df['Leather_interior'] = df['Leather_interior'].astype(str)
df['Fuel_type'] = df['Fuel_type'].astype(str)
df['Gear_box_type'] = df['Gear_box_type'].astype(str)
df['ProdYear'] = df['ProdYear'].astype(str)
```

```python
# check for white spaces in string values in all columns
for col in df.columns:
    # convert non-string values to string
    mask = df[col].apply(lambda x: str(x)).str.contains(r'^\s|\s$')
    # count the number of values with white spaces
    num_spaces = mask.sum()
    if num_spaces > 0:
        print(f'Column "{col}" contains {num_spaces} values with white spaces.')
```
