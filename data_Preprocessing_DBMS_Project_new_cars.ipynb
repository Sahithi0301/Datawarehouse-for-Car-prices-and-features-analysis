```python
# Split Car Full Data into Brand, Model, And Year of The Model

model = df['Car Model'].apply(lambda x: ''.join((x.split()[1],' ',x.split()[2])) if len(x.split()) >4 else x.split()[1])
year = df['Car Model'].apply(lambda x: x.split()[-1])
brand = df['Car Model'].apply(lambda x: x.split()[0])


df['Manufacturer'], df['Model'], df['ProdYear'] = brand,model,year
```

```python


# Define a function to classify transmission type
def classify_transmission(model):
    automatic_keywords = ['A/T', 'Automatic', 'Automtic','AT']
    manual_keywords = [ 'Manual\u200f', 'manual\u200f','Manual', 'M/T']

    for keyword in automatic_keywords:
        if keyword in model:
            return 'Automatic'
    for keyword in manual_keywords:
        if keyword in model:
            return 'Manual'
    return 'Unknown'


# Create a new column 'transmission' based on the 'model' column
df['Gear_box_type'] = df['Car Model'].apply(classify_transmission)

print(df)
```

```python

Price_variation=df['Price Change'].apply(lambda x: x.split()[0])

Price_change=df['Price Change'].apply(lambda x: x.split()[1])

df['Price_variation'], df['Price_change']= Price_variation,Price_change
```

```python
# Convert Average price, Minimum price, Maximum price to Numerical Values

for i in ['Old Price','New Price','Price_change']:
    df[i] = df[i].apply(lambda x : float((str(x).replace(',','')).split()[0]))

df['Price_change'] = df['Price_change'].abs()
```
